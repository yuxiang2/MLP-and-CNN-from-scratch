```python
import numpy as np
import functions as F
import module as M
import data as D
import matplotlib.pyplot as plt
```

```python
# Load the data
train = {}
dev = {}
test = {}

train['data'] = np.load('data/train_img.npy')
train['label'] = np.load('data/train_label.npy')

dev['data'] = np.load('data/dev_img.npy')
dev['label'] = np.load('data/dev_label.npy')

test['data'] = np.load('data/test_img.npy')
test['label'] = np.load('data/test_label.npy')
```

```python
num_class = 10
```

### Q1

```python
model = M.ListModule([
    
    M.Conv2d(3, 36, 5, 1, 2), 
    M.Maxpool2d(2),
    F.ReLU(), 
    
    M.Flatten(),
    M.Linear(9216, 10),
])
loss_f = F.SoftmaxEntropy(num_class)
```

```python
train_loss = []
train_acc = []
dev_loss = []
dev_acc = []
test_loss = []
test_acc = []
```

```python
for e in range(50):
    loss, acc = D.trainEpoch(model, loss_f, train['data'], train['label'], 0.0005, momentum=0.9, batch_size=32)
    train_loss.append(loss)
    train_acc.append(acc)

    loss, acc = D.trainEpoch(model, loss_f, dev['data'], dev['label'], 0.0005, batch_size=32, train=False)
    dev_loss.append(loss)
    dev_acc.append(acc)

    loss, acc = D.trainEpoch(model, loss_f, test['data'], test['label'], 0.0005, batch_size=32, train=False)
    test_loss.append(loss)
    test_acc.append(acc)

    print('e={}'.format(e),end='\r')
```

```python
plt.figure(figsize=(15,10))
plt.plot(train_loss, color='b', label='train loss')
plt.plot(dev_loss, color='r', label='validation loss')
plt.plot(test_loss, color='g', label='test loss')
plt.xlabel('Number Epoches')
plt.ylabel('Avg Cross Entropy Loss')
plt.title('Loss VS. Epoches')
plt.legend()
plt.show()
```

```python
print(train_acc[-1])
print(dev_acc[-1])
print(test_acc[-1])
```
